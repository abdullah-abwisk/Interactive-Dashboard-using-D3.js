```python
import csv
import json

class Node(object):
    def __init__(self, name, value=None, extra=None):
        self.name = name
        self.children = []
        self.value = value
        self.extra = extra

    def child(self, cname, value=None, extra=None):
        child_found = [c for c in self.children if c.name == cname]
        if not child_found:
            _child = Node(cname, value, extra)
            self.children.append(_child)
        else:
            _child = child_found[0]
        return _child
    def as_dict(self):
        res = {'name': self.name}
        if self.value is None:
            res['children'] = [c.as_dict() for c in self.children]
        else:
            res['value'] = self.value

        if self.extra:
            res.update(self.extra)

        return res

root = Node('United States')

with open("../Dataset/hierarchy.csv", 'r') as f:
    reader = csv.reader(f)
    header = next(reader)
    for row in reader:
        Country, OriginState, AirportName, FlightDate = row
        print(Country, OriginState, AirportName, FlightDate)
        root.child(OriginState) \
            .child(AirportName,value=FlightDate) \

print (json.dumps(root.as_dict(), indent=4))
with open("../Dataset/hier.json", 'w', encoding='utf-8') as jsonf:
    jsonf.write(json.dumps(root.as_dict(), indent=4))
```
